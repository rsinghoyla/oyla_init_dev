```python
import pickle as pkl
import numpy as np
```

```python
def cross_counting(results, key1, key2):
    person_cross_counts={}
    person_cross_counts['found_none'] = []
    person_cross_counts['found_both'] = []
    person_cross_counts['found_more_one'] = []
    person_cross_counts['found_more_two'] = []
    for k in sorted(results[key1].keys()):
        flag1 = 0
        for i in range(len(results[key1][k])):
            if 'person' in results[key1][k][i][0]:
                flag1 += 1
        flag2 = 0
        for i in range(len(results[key2][k])):
            if 'person' in results[key2][k][i][0]:
                flag2 += 1
        if flag1 == flag2 and flag1 == 0:
            person_cross_counts['found_none'].append(k+' '+str(flag1)+' '+str(flag2))
        if flag1 == flag2 and flag1>0:
            person_cross_counts['found_both'].append(k+' '+' '+str(flag1)+' '+str(flag2))
        if flag1>0 and flag1>flag2:
            person_cross_counts['found_more_one'].append(k+' '+str(flag1)+' '+str(flag2))
        if flag2>0 and flag2>flag1:
            person_cross_counts['found_more_two'].append(k+' '+str(flag1)+' '+str(flag2))
    l = 0
    for k in person_cross_counts.keys():
        l += len(person_cross_counts[k])
        print(k, len(person_cross_counts[k]))
    print(l)
    return person_cross_counts
```
