```python
import sklearn.neighbors
```

```python
(ii,jj) =np.where(thresholded_ampl<65000)
ta = thresholded_ampl[ii,jj]
fp = filtered_phase[ii,jj]
```

```python
features = np.vstack((ii,jj,ta)).transpose()
```

```python
A = sklearn.neighbors.kneighbors_graph(features, 100,mode='distance',n_jobs=100)
```

```python
from scipy.cluster.vq import whiten
features_w = whiten(features)
```

```python
B = sklearn.neighbors.kneighbors_graph(features_w, 100,mode='distance',n_jobs=100)
```

```python
print(scipy.sparse.find(A[289,:])[1])
print(scipy.sparse.find(B[289,:])[1])
```

```python
print(features[289])
print(features[139])
print(features[172])
print(features_w[289])
print(features_w[139])
print(features_w[172])
```

```python
np.std(features_w,axis=0)
```

```python
def graph_median(i):
    return np.median(fp[scipy.sparse.find(B[i,:])[1]])
```

```python
r = []
import time
start = time.time()
for i in range(A.shape[0]):
    r.append(graph_median(i))
end = time.time()
print(end-start)
```

```python
from joblib import Parallel, delayed
import multiprocessing
my_list = range(A.shape[0])
start = time.time()
results = []
num_cores = multiprocessing.cpu_count()
results = Parallel(n_jobs=num_cores, verbose=0)(delayed(
    graph_median)(i)for i in my_list)
end = time.time()
print(end-start)
```

```python
assert r==results
```
