```python
trimap = dist.copy()
trimap = trimap.astype('float32')
trimap[ampl>65300] = 0.5
trimap[ampl<65300] = 1
trimap[ampl==65300] = 0
trimap[ampl<=100] = 0.0
```

```python
I = foreground
for i in range(3):
    I[:,:,i] = alpha*foreground[:,:,i]+(1-alpha)*background[:,:,i]
```
