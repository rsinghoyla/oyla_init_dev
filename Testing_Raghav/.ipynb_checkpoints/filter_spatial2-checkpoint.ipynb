Comparing median filter and pcl median filter for different filter sizes. Default pcl median is 100 and its most aggresive of lot. Default median is 3 iterations of 3x3 and least aggresive of lot. PCL median at 49 is better than median at 1 iteration and 7x7. In fact median 7x7 ==49 has worst performance. There are five versions of Pcl median 100, 49, 49 with division by np.max(ii), 49 with whitening and radius search with neighbour on. MEDIAN + EDGE SEEMS to be doing as well as others
